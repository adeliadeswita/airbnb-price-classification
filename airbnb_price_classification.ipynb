Akomodasi merupakan salah satu kebutuhan utama yang perlu diperhatikan ketika bepergian ke luar negeri. Dataset yang tersedia berisikan sebanyak 796 records pemesanan AirBnb yang berada di kawasan Amsterdam dengan 4 jenis kategori harga (Low Price: EUR10-EUR100; Medium Price: EUR100-EUR250; Mid-High Price: EUR250-EUR350; Highest Price: >EUR350).

# Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,:7]
y = df.iloc[:,7]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30, random_state=42)
```

# Data Modelling

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model = model.fit(x_train, y_train)
print(model.predict(x_test))
print(y_test)
```

# Evaluation

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = model.predict(x_test)
confusion_matrix(y_pred, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
